### Problem statement

The data is copied from the linear regression exercise of machine learning course at https://www.coursera.org/course/ml
The file ex1data2.txt contains a training set of housing prices in Portland, Oregon. The first column is the 
size of the house (in square feet), the second column is the number of bedrooms, 
and the third column is the price of the house, which we want to predict.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = np.loadtxt("ex2data2.txt", delimiter=",")
    X_train = data[:,:2]
    Y_Train = data[:,2]
    return X_train,Y_Train
```

```python
X_train, Y_train = load_data()
print(X_train)
print(Y_train)
X_features = np.array(["Size(Sqft)","Bedrooms"])
```

```python
## As the features are showing different ranges we need to scale
scaler = StandardScaler()
X_norm = scaler.fit_transform(X_train)
linear_model = LinearRegression()
linear_model.fit(X_norm,Y_train)
```

```python
print(f"Coeff = {linear_model.coef_} intercept = {linear_model.intercept_}")
```

```python
## Predict for x_norm
Y_lm_predicted = linear_model.predict(X_norm)
print(Y_lm_predicted)
print(Y_train)
```

```python
fig,ax = plt.subplots(1,2,figsize=(12,4))
for i in range(len(ax)):
    ax[i].scatter(X_norm[:,i],Y_train, label="Train")
    ax[i].set_xlabel(X_features[i])
    ax[i].set_ylabel("Price")
    ax[i].scatter(X_norm[:,i],Y_lm_predicted,color="Orange", label="Predict")
    ax[i].legend()
```
