#### Support Vector Regression

* Works on priciple of classificaiton model of SVM
* Works well for non-linear training set


```python
# Read data  --> feature scale --> SVR fit 
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data():
    data = pd.read_csv("Position_Salaries.csv")
    print(data)
    X = data.iloc[:,1:-1].values
    Y = data.iloc[:,-1].values
    return X,Y
```

```python
X,Y = load_data()
```

```python
## Plot x & y
plt.scatter(X,Y)
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

Y = Y.reshape(-1,1) # to reshape to 2D

x= sc_x.fit_transform(X)
y = sc_y.fit_transform(Y)
```

```python
## fit in SVR
from sklearn.svm import SVR

svr_reg = SVR(kernel='rbf')
svr_reg.fit(x,y)
```

```python
sc_y.inverse_transform(svr_reg.predict(sc_x.transform([[6.5]])).reshape(-1,1))
```

```python
svr_reg.predict(sc_x.transform([[6.5]]))
```

```python
plt.scatter(X,Y)
plt.plot(X,sc_y.inverse_transform(svr_reg.predict(x).reshape(-1,1)), color="red")
plt.show()
```
