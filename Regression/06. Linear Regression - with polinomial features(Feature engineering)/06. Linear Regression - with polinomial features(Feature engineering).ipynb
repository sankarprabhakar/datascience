#### Problem statement
Given the dataset of positions and salary, we are trying to come up with a model to predict the salary given the position 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## Load the data and visualize
def load_data():
    data = pd.read_csv("Position_Salaries.csv")
    print(data)
    X = data.iloc[:,1:-1].values
    Y = data.iloc[:,-1].values
    return X,Y
    
    
```

```python
X,Y = load_data()
```

```python
## Plot x & y
plt.scatter(X,Y)
plt.show()
```

```python
## to fit a linear model
from sklearn.linear_model import LinearRegression
lin_reg1 = LinearRegression()
lin_reg1.fit(X,Y)
print(f"Coef = {lin_reg1.coef_} intercept = {lin_reg1.intercept_}")
```

```python
## Try to predict using lin_reg1
y_pred = lin_reg1.predict(X)
```

```python
## Plot x & y
plt.scatter(X,Y)
plt.plot(X,y_pred)
plt.show()
```

```python
# From the figure it is clear we are not able to fit the graph. so we will try to do
#feature engineering to generate polinomial features

from sklearn.preprocessing import PolynomialFeatures
pol_feature = PolynomialFeatures(degree=4)
x_poly = pol_feature.fit_transform(X)

lin_reg2 = LinearRegression()
lin_reg2.fit(x_poly,Y)
print(f"Coef = {lin_reg2.coef_} intercept = {lin_reg2.intercept_}")
```

```python
## Try to predict using lin_reg1
y_poly_pred = lin_reg2.predict(pol_feature.fit_transform(X))
```

```python
## Plot x & y
plt.scatter(X,Y)
plt.plot(X,y_poly_pred)
plt.show()
```
