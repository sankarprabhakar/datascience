### Problem statment
Experience of person and Salary is given as training set.  We will fit the model to predict salary based on the experience.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#### Note: Dataframe.values gives a Numpy type of 2D array. Each row is represented as array

```python
def load_data():
    data = pd.read_csv("Salary_Data.csv")
    #print(data)
    X_train = data.iloc[:,:-1].values 
    Y_train = data.iloc[:,-1].values
    return X_train, Y_train
```

```python
X_train, Y_train = load_data()
print(X_train)
print(Y_train)
```

```python
plt.scatter(X_train,Y_train)
plt.xlabel("Experience")
plt.ylabel("salary")
plt.title("Exp Vs Salary")
plt.show()
```

```python
### split test set into train and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_train,Y_train, random_state=0, test_size=1/3)
```

```python
print(len(x_train) , len(x_test))
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x_train,y_train)
```

```python
print(f"coef = {linear_reg.coef_} intercept = {linear_reg.intercept_}")
```

```python
## Predict output
y_predict = linear_reg.predict(x_train)
```

```python
plt.scatter(x_train,y_train, color="red")
plt.plot(x_train,y_predict)
plt.xlabel("Experience")
plt.ylabel("salary")
plt.title("Exp Vs Salary")
plt.show()
```

```python
## Predict for test set
## Predict output
y_test_predict = linear_reg.predict(x_test)
```

```python
plt.scatter(x_test,y_test, color="red")
plt.plot(x_test,y_test_predict)
plt.xlabel("Experience")
plt.ylabel("salary")
plt.title("Exp Vs Salary")
plt.show()
```
