## Problem statement

Given a list of training set of purchase history from social media ad, age of a person and income. Come up with a  model give the age and income predeict if he/she will purchase a product based on social media ad

```python
import numpy as np
import pandas as pd
```

```python
def load_data():
    data = pd.read_csv("Social_Network_Ads.csv")
    #print(data)
    X = data.iloc[:,:-1].values ## Give out put as ndarry
    Y = data.iloc[:,-1].values
    #print(X)
    #print(Y)
    return X,Y
                       
```

```python
X,Y = load_data()
```

```python
Y[0:5]
```

```python
## As the features are very divergent do feature scaling
from sklearn.preprocessing import StandardScaler

scalar = StandardScaler()
X = scalar.fit_transform(X)
print(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, random_state=0, test_size=1/3)
```

```python
## Split data into train and test set
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
## Predict result 
Y_pred = log_reg.predict(X_test)
        
```

```python
result = np.concatenate((Y_test.reshape(-1,1), Y_pred.reshape(-1,1)),1)
#print(result)
```

```python
log_reg.score(X_test,Y_test)
```

```python
## Getting the confusion matrix and accuracy score
from sklearn.metrics import confusion_matrix, accuracy_score

confusion_matrix(Y_test,Y_pred)
```

```python
## print accurary score - Same as log_reg.score
accuracy_score(Y_test, Y_pred)
```
