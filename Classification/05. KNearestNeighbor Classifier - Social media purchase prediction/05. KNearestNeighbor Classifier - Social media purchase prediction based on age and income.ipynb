## Problem statement

Given a list of training set of purchase history from social media ad, age of a person and income. Come up with a  model give the age and income predeict if he/she will purchase a product based on social media ad

```python
import numpy as np
import pandas as pd
```

```python
def load_data():
    data = pd.read_csv("Social_Network_Ads.csv")
    #print(data)
    X = data.iloc[:,:-1].values ## Give out put as ndarry
    Y = data.iloc[:,-1].values
    #print(X)
    #print(Y)
    return X,Y
                       
```

```python
X,Y = load_data()
```

```python
## Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
## Split into training and test set
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.25, random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_classifier = KNeighborsClassifier(metric='minkowski', n_neighbors=5, p=2)
```

```python
knn_classifier.fit(X_train, Y_train)
```

```python
## Predict result
Y_pred = knn_classifier.predict(X_test)
```

```python
knn_classifier.score(X_test,Y_test)
```

```python
## Calculate the confusion matrix and accuracy 

from sklearn.metrics import accuracy_score, confusion_matrix

confusion_matrix(Y_test, Y_pred)
```

```python
accuracy_score(Y_test,Y_pred)
```
