# Linear regression modelling - Concept
Problem statement:
Given the size of house and the actual prices it is sold for. A project given by the real estate agent is to come up with basic model which can predict the price of a house given its size.

For building up the concept of linear regression let us consider we only have 3 data points. Prices of 1000 sqft house is 300000 dollars. 2000sqft house cost 500000 dollars and 3000 sqft house cost 700 dollars. Given this data how to predict the cost of 1200 sqft house.

| Size (1000 sqft)     | Price (1000s of dollars) |
| -------------------| ------------------------ |
| 1.0               | 300                      |
| 2.0               | 500                      |
| 3.0               | 700                      |



```python
# importing numpy and pandas
import numpy as np
import pandas as pd
x_train = np.array([1.0,2.0,3.0])
y_train = np.array([300,500,700])
print(f"shape of x_train = {x_train.shape}")
print(f"shape of x_train = {y_train.shape}")
```

Note:  Total number of training examples is denoted by 'm'. Here in the above example it is 3. we can get len() to get number of training.

One entry in the training set is denoted by x$^{(i)}$,y$^{(i)}$ . In the example x$^{1}$ is 1 and y$^{1}$ is 300;  x$^{2}$ is 2 and y$^{2}$ is 500 and so on and so forth 

```python
# Length of training set
print(f"Lenght of training set = {len(x_train)}")
```

### Plotting the graph of given training set.
We can use the mathplotlib to plot the training set in a scatter plot


```python
from matplotlib import pyplot as plt
plt.scatter(x_train,y_train, marker='x',c='b')
plt.title("House size vs Price")
plt.xlabel("Size in 1000 of Sqft")
plt.ylabel("Price of house in 1000 dollars")
plt.show()
```

### Model representation
For linear regression with  x$^{(i)}$,y$^{(i)}$, the model would be represented by below formula

$$ f_{w,b}(x^{(i)}) = wx^{(i)} + b \tag{1}$$

this is formula for straight line. By choosing different values of w & b we can get a line which is closer to the available inputs

```python
import ipywidgets as widgets
print("Choose a value of w & b")
w_slider  = widgets.IntSlider(
    min=0,
    max=1000,
    step=50,
    description='w value:',
    value=100
)
display(w_slider)
b_slider  = widgets.IntSlider(
    min=0,
    max=1000,
    step=50,
    description='b value:',
    value=100
)
display(b_slider)
```

Note : Let us define a function which calculates the predicted cost for a given w, b and input

```python
def compute_predicted_cost(x,w,b):
    '''x is a input training set
     w -> Paremeter
     b -> base
     '''
    m = x.shape[0] # first element in the tuple gives the number of training elements
    predicted_output = np.zeros(m)
    for i in range(0,m):
        predicted_output[i] = w * x[i] + b
    return predicted_output
```

Note: Let us try to input the training set and plot the output for different values of w & b. and see which line better fits the given input

```python
predicted_value = compute_predicted_cost(x_train,w_slider.value, b_slider.value)
plt.scatter(x_train,y_train, marker='x',c='b',label='Our Prediction')
plt.plot(x_train,predicted_value, c='r',label='Actual value')

plt.title("House size vs Price")
plt.xlabel("Size in 1000 of Sqft")
plt.ylabel("Price of house in 1000 dollars")
plt.legend()
plt.show()
```

Note : For value of w = 200 and b =100 , the predicted line is exactly fitting the given line. We figured out this line with a lot of experiments. *But there are mathematical ways to get the value of w & b for a given training set.*

Now let us try to predict the value of 1200sqft house

```python
new_house_size = 1.2
w=200
b=100
predicted_cost = w*new_house_size + b
print(f"Predicted cost = {predicted_cost} thousand dollars")
```
